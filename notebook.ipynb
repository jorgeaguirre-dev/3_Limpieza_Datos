## Limpieza

```python
# Elimino columnas con casi todo nulos
df1 = df.drop(columns=['Unnamed: 21', 'Unnamed: 22'])
```

Veamos la cantidad de nulos por campo.

```python
df1.isnull().sum().sort_values(ascending=False)
```

Se observan campos con gran cantidad de nulos, aunque no serían los de mayor interés para el análisis actual.
Seguidamente analizamos porcentualmente la incidencia de nulos para saber el impacto en el error que pudiera introducir debido a información faltante.

```python
# Porcentaje de nulos por columna
porcentaje_nulos = (df1.isnull().sum() / len(df1)) * 100
print(porcentaje_nulos.sort_values(ascending=False))
```

Afortunadamente, el impacto en los campos de mayor interés (Country y Type) es mínimo.

### Campo Year

```python
# Convertir a entero 
df1['Year'] = df1['Year'].astype('Int64')
df1[['Year']].head()
```

Ahora analizamos posibles años extraños o cargados con error.

```python
df1[df1['Year'] < 1500][['Country','Type','Location','Year']]
```

### Campo Country
Será necesario eliminar los Country='NaN'

```python
# Borrar filas donde la columna Country sea nula
df1.dropna(subset=['Country'], inplace=True)
```

De igual manera procedemos con campo 'Type' cuya baja incidencia de NaN lo hace posible.

```python
# Quitar espacios al en los extremos y poner primera letra en mayúsculas
df1['Country'] = df1['Country'].str.strip().str.capitalize()
```

```python
# Vemos los 20 Country más frecuentes
print(df1['Country'].value_counts().head(20))
```

### Campo Type
La tabla top anterior luce bien en cuanto a los principales países.

```python
# Borrar filas donde la columna Type sea nula
df1.dropna(subset=['Type'], inplace=True)
```

```python
# Quitar espacios al en los extremos y poner primera letra en mayúsculas
df1['Type'] = df1['Type'].str.strip().str.capitalize()
```

```python
# Vemos los Type más frecuentes
print(df1['Type'].value_counts().head(20))
```

Consideramos Boat como Watercraft

```python
# Unificar Boat en Watercraft
df1['Type'] = df1['Type'].replace('Boat', 'Watercraft')
```

Ahora los casos con frecuencia 1 y etiquetas dudosas

```python
# Contamos las frecuencias
counts = df1['Type'].value_counts()
```

```python
counts
```

Eliminamos los casos de una sola aparición o irrelevantes

```python
# Lista de tipos que queremos eliminar 
tipos_a_borrar = ['?', 'Unconfirmed', 'Unverified', 'Under investigation']

# Filtrar: Nos quedamos con todo lo que NO esté en esa lista
df = df1[~df1['Type'].isin(tipos_a_borrar)]

# Verificamos el resultado final
print(df['Type'].value_counts())
```

### Campos Age, Sex, Fatal

```python
# Los campos de interés
cols_to_fix = ['Sex', 'Fatal Y/N', 'Age']

# Limpiar espacios y pasar a mayúsculas
for col in cols_to_fix:
    df[col] = df[col].astype(str).str.strip().str.upper()
```

```python
df['Sex'].value_counts()
```

```python
# Imputación directa (M X 2 y N pasan a M)
df['Sex'] = df['Sex'].replace(['M X 2', 'N'], 'M')
```

```python
# Imputación directa del resto
df['Sex'] = df['Sex'].replace(['LLI'], 'F')
df['Sex'] = df['Sex'].replace(['?','.'], 'M')
```

```python
df['Sex'].value_counts()
```

Ahora Age

```python
print(df['Age'].value_counts().to_string())
```

```python
# Limpieza básica
from clean_age import clean_age

df['Age2'] = df['Age'].apply(clean_age)
```

```python
print(df['Age2'].value_counts().to_string())
```

Si bien ha mejorado mucho, no se pudieron estandarizar todos los casos todavía, aunque no es un campo principal para nuestro análisis está claro que tiene muchos errores de carga manual y falta de criterio.

## Cantidad de Incidentes por Tipo y País

```python
# Agrupamos por País y Tipo de ataque
df_tipo_pais = df.groupby(['Country', 'Type']).size().reset_index(name='Incident_Count').sort_values(by='Incident_Count', ascending=False)
df_tipo_pais.head(10)
```

```python
# Pivot
tabla_pivot = df_tipo_pais.pivot(index='Country', columns='Type', values='Incident_Count').fillna(0)

display(tabla_pivot)
```

```python
# Creamos un pivot table
tabla_pivot = df_tipo_pais.pivot(index='Country', columns='Type', values='Incident_Count').fillna(0)

# Calculamos la columna 'Total' sumando todas las columnas de tipos
tabla_pivot['Total'] = tabla_pivot.sum(axis=1)

# Ahora ordenamos por la columna 'Total'
tabla_pivot = tabla_pivot.sort_values(by='Total', ascending=False)

# Convertimos a entero para mejor visual
tabla_pivot = tabla_pivot.astype(int)

display(tabla_pivot.head(10))
```

## Exportación de resultados

```python
# Preparar para exportar
tabla_final_export = tabla_pivot.reset_index()
tabla_final_export.to_csv('mapa_ataques_tiburones.csv', index=False)
```

## Visualización de Resultados

```python
fig = px.choropleth(tabla_final_export, 
                    locations="Country", 
                    locationmode='country names',
                    color="Total", 
                    hover_name="Country", 
                    # Usamos una escala de colores rojos
                    color_continuous_scale=px.colors.sequential.Reds,
                    title='Distribución Global de Ataques de Tiburón (Total)')

fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
fig.show()
```
